```python
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Assume you already have X_train, X_test, y_train, y_test, and df

# Creating a list of feature labels (feat_labels) excluding 'Feeling anxious'
feat_labels = df.drop(columns=['Feeling anxious']).columns

# X is all columns except 'Feeling anxious'
X = df.drop(columns=['Feeling anxious']).values

# y is the column 'Feeling anxious'
y = df['Feeling anxious'].values

# Splitting into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)

# Initializing XGBoost model
xgb_model = xgb.XGBClassifier(n_estimators=1000, random_state=1, n_jobs=-1)

# Training the model on the training data
xgb_model.fit(X_train, y_train)

# Evaluating feature importance
importances = xgb_model.feature_importances_
indices = np.argsort(importances)[::-1]

# Displaying feature importance
for f in range(X_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30, feat_labels[indices[f]], importances[indices[f]]))

# Plotting feature importance using matplotlib
plt.figure(figsize=(10, 6))
plt.bar(range(X_train.shape[1]), importances[indices], align='center')
plt.xticks(range(X_train.shape[1]), feat_labels[indices], rotation=90)
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Feature Importance')
plt.tight_layout()
plt.show()

# Evaluating the model accuracy on the test data
y_pred = xgb_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Model Accuracy on Test Data: {accuracy:.4f}')
```
